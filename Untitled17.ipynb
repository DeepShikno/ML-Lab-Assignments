```python
# =============================================================================
# ML Lab Assignment - Employee Attrition Prediction
# Dataset: employee_attrition_Mix.csv
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (accuracy_score, confusion_matrix,
                              classification_report, ConfusionMatrixDisplay)

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

# =============================================================================
# STEP 1: Load Dataset
# =============================================================================
df = pd.read_csv('employee_attrition_Mix.csv')

print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)
print(f"Shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nTarget Distribution:")
print(df['Attrition'].value_counts())

# =============================================================================
# STEP 2: Preprocessing
# =============================================================================
print("\n" + "=" * 60)
print("PREPROCESSING")
print("=" * 60)

# Label Encode target: No -> 0, Yes -> 1
le = LabelEncoder()
df['Attrition'] = le.fit_transform(df['Attrition'])

# Features and Target
X = df.drop('Attrition', axis=1)
y = df['Attrition']

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split (80-20, stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Training samples : {X_train.shape[0]}")
print(f"Testing  samples : {X_test.shape[0]}")

# =============================================================================
# STEP 3: Train Multiple Models
# =============================================================================
models = {
    'Logistic Regression' : LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree'       : DecisionTreeClassifier(random_state=42),
    'Random Forest'       : RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting'   : GradientBoostingClassifier(random_state=42),
    'SVM'                 : SVC(random_state=42, probability=True)
}

results = {}

print("\n" + "=" * 60)
print("MODEL TRAINING & EVALUATION")
print("=" * 60)

for name, model in models.items():
    # Train
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Metrics
    acc    = accuracy_score(y_test, y_pred)
    cm     = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred,
                                   target_names=['No', 'Yes'],
                                   output_dict=True)

    results[name] = {
        'model' : model,
        'y_pred': y_pred,
        'acc'   : acc,
        'cm'    : cm,
        'report': report
    }

    print(f"\n{'─' * 50}")
    print(f"Model: {name}")
    print(f"Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred, target_names=['No', 'Yes']))

# =============================================================================
# STEP 4: Summary Comparison Table
# =============================================================================
print("\n" + "=" * 60)
print("MODEL COMPARISON SUMMARY")
print("=" * 60)

summary = []
for name, r in results.items():
    rep = r['report']
    summary.append({
        'Model'           : name,
        'Accuracy'        : round(r['acc'], 4),
        'Precision (Yes)' : round(rep['Yes']['precision'], 4),
        'Recall (Yes)'    : round(rep['Yes']['recall'],    4),
        'F1-Score (Yes)'  : round(rep['Yes']['f1-score'],  4),
    })

df_summary = pd.DataFrame(summary).sort_values('Accuracy', ascending=False)
print(df_summary.to_string(index=False))

best_model = df_summary.iloc[0]['Model']
print(f"\n>>> Best Model: {best_model} <<<")

# =============================================================================
# STEP 5: Visualizations
# =============================================================================

# --- Plot 1: Confusion Matrices ---
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, (name, r) in enumerate(results.items()):
    disp = ConfusionMatrixDisplay(confusion_matrix=r['cm'],
                                  display_labels=['No', 'Yes'])
    disp.plot(ax=axes[i], colorbar=False, cmap='Blues')
    axes[i].set_title(name, fontsize=12, fontweight='bold')

axes[5].set_visible(False)
plt.suptitle('Confusion Matrices - All Models', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()

# --- Plot 2: Accuracy Comparison ---
fig, ax = plt.subplots(figsize=(9, 5))
colors = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0', '#F44336']

bars = ax.barh(
    [r['Model'] for r in summary],
    [r['Accuracy'] for r in summary],
    color=colors, height=0.5
)

ax.set_xlabel('Accuracy', fontsize=12)
ax.set_title('Model Accuracy Comparison', fontsize=14, fontweight='bold')
ax.set_xlim(0, 1.1)

for bar, val in zip(bars, [r['Accuracy'] for r in summary]):
    ax.text(bar.get_width() + 0.005, bar.get_y() + bar.get_height() / 2,
            f'{val:.4f}', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('accuracy_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

# --- Plot 3: Precision / Recall / F1 Comparison ---
fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(len(summary))
w = 0.25

ax.bar(x - w, [r['Precision (Yes)'] for r in summary], w,
       label='Precision', color='#2196F3')
ax.bar(x,     [r['Recall (Yes)']    for r in summary], w,
       label='Recall',    color='#4CAF50')
ax.bar(x + w, [r['F1-Score (Yes)']  for r in summary], w,
       label='F1-Score',  color='#FF9800')

ax.set_xticks(x)
ax.set_xticklabels([r['Model'] for r in summary], rotation=15, ha='right')
ax.set_ylim(0, 1.15)
ax.set_title('Precision / Recall / F1-Score for Attrition = Yes',
             fontsize=13, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.savefig('metrics_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```
